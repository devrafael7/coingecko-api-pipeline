```sql
CREATE OR REPLACE TABLE coingecko.silver.crypto_markets AS
SELECT DISTINCT
    id,
    symbol,
    name,

    CAST(current_price AS DOUBLE) AS current_price,
    CAST(total_volume AS BIGINT) AS total_volume,
    CAST(market_cap AS BIGINT) AS market_cap,
    CAST(market_cap_rank AS INT) AS market_cap_rank,

    CAST(high_24h AS DOUBLE) AS high_24h,
    CAST(low_24h AS DOUBLE) AS low_24h,

    CAST(price_change_24h AS DOUBLE) AS price_change_24h,
    CAST(price_change_percentage_24h AS DOUBLE) AS price_change_percentage_24h,

    TO_TIMESTAMP(atl_date) AS atl_date,
    CAST(atl AS DOUBLE) AS atl,
    CAST(ath AS DOUBLE) AS ath,
    CAST(ath_change_percentage AS DOUBLE) AS ath_change_percentage,

    TO_TIMESTAMP(last_updated) AS last_updated,
    CAST(ingestion_date AS DATE) AS ingestion_date,

    CURRENT_TIMESTAMP() AS silver_processed_at

FROM coingecko.bronze.crypto_markets_raw
WHERE id IS NOT NULL
  AND symbol IS NOT NULL
  AND name IS NOT NULL;
```
