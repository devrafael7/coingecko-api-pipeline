```sql
CREATE OR REPLACE TABLE coingecko.gold.market_overview AS
SELECT
    ingestion_date,
    COUNT(*) AS total_coins,
    SUM(market_cap) AS total_market_cap,
    SUM(total_volume) AS total_volume_24h,
    AVG(current_price) AS avg_current_price,
    MAX(current_price) AS max_current_price,
    MIN(current_price) AS min_current_price
FROM coingecko.silver.crypto_markets
GROUP BY ingestion_date;
```

```sql
CREATE OR REPLACE TABLE coingecko.gold.top_10_by_market_cap AS
SELECT
    market_cap_rank,
    id,
    symbol,
    name,
    current_price,
    market_cap,
    total_volume,
    price_change_percentage_24h,
    last_updated,
    ingestion_date
FROM coingecko.silver.crypto_markets
WHERE market_cap_rank <= 10;
```

```sql
CREATE OR REPLACE TABLE coingecko.gold.price_performance_24h AS
SELECT
    id,
    symbol,
    name,
    current_price,
    price_change_24h,
    price_change_percentage_24h,
    market_cap,
    total_volume,
    CASE
        WHEN price_change_percentage_24h > 0 THEN 'positive'
        WHEN price_change_percentage_24h < 0 THEN 'negative'
        ELSE 'neutral'
    END AS performance_status,
    ingestion_date
FROM coingecko.silver.crypto_markets
ORDER BY price_change_percentage_24h DESC;
```
